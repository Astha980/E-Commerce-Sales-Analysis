```python
import pandas as pd
orders = pd.read_csv("olist_orders_dataset.csv")
order_items = pd.read_csv("olist_order_items_dataset.csv")
products = pd.read_csv("olist_products_dataset.csv")
reviews = pd.read_csv("olist_order_reviews_dataset.csv")
customers = pd.read_csv("olist_customers_dataset.csv")
category = pd.read_csv("product_category_name_translation.csv")
```

```python
orders.info()
orders.head()
products.info()
products.head()
reviews.info()
reviews.head()
customers.info()
customers.head()
category.info()
category.head()
order_items.info()
order_items.head()
```

```python
orders['order_purchase_timestamp'] = pd.to_datetime(
    orders['order_purchase_timestamp']
)
print(orders['order_purchase_timestamp'].dtype)
```

```python
print(orders.isnull().sum())
print(order_items.isnull().sum())
print(products.isnull().sum())
print(reviews.isnull().sum())
print(customers.isnull().sum())
print(category.isnull().sum())
```

```python
orders.drop_duplicates(inplace=True)
order_items.drop_duplicates(inplace=True)
products.drop_duplicates(inplace=True)
reviews.drop_duplicates(inplace=True)
customers.drop_duplicates(inplace=True)
category.drop_duplicates(inplace=True)
```

```python
products = products.merge(
    category,
    on='product_category_name',
    how='left'
)
```

```python
df=orders.merge(
    customers,
    on='customer_id',
    how='left'
)
```

```python
df = df.merge(
    order_items,
    on='order_id',
    how='left'
)
```

```python
df = df.merge(
    products,
    on='product_id',
    how='left'
)
```

```python
df = df.merge(
    reviews,
    on='order_id',
    how='left'
)
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df['Month'] = df['order_purchase_timestamp'].dt.to_period('M')
print(df['Month'])
```

```python
#Which product category has the highest revenue?
category_revenue = (
    df.groupby('product_category_name')['price']
      .sum()
      .sort_values(ascending=False)
)

print(category_revenue.head(10))
```

```python
#Which month had peak sales?
monthly_sales = (
    df.groupby('Month')['price']
      .sum()
)

print(monthly_sales)
```

```python
#Which region performs best?
regional_sales = (
    df.groupby('customer_state')['price']
      .sum()
      .sort_values(ascending=False)
)

print(regional_sales)
```

```python
# What is the average order value trend?
order_value = (
    df.groupby('order_id')['price']
      .sum()
      .reset_index()
)

average_order_value = order_value['price'].mean()

print(average_order_value)
```

```python
# for monthly trend
monthly_avg = (
    df.groupby('Month')['price']
      .mean()
)

print(monthly_avg)
```

```python
#What is the customer review score distribution?
review_distribution = df['review_score'].value_counts().sort_index()
print(review_distribution)
```

```python
# Revenue by Category
import matplotlib.pyplot as plt
category_revenue.head(10).plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Product Categories by Revenue")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
monthly_sales.plot(kind='line', marker='o', figsize=(10,5))
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
regional_sales.head(10).plot(kind='bar')
plt.title("Top Regions by Revenue")
plt.xlabel("State")
plt.ylabel("Revenue")
plt.show()
```

```python
order_value['price'].plot(kind='hist', bins=30)
plt.title("Order Value Distribution")
plt.xlabel("Order Value")
plt.show()
```

```python
review_distribution.plot(kind='pie', autopct='%1.1f%%')
plt.title("Review Score Distribution")
plt.ylabel("")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
pivot = df.pivot_table(
    values='price',
    index='product_category_name_english',
    columns='Month',
    aggfunc='sum',
    fill_value=0
)
```

```python
plt.figure(figsize=(14, 12))
sns.heatmap(
    pivot,
    cmap="YlGnBu")
plt.title("Revenue by Product Category and Month")
plt.xlabel("Month")
plt.ylabel("Product Category")
plt.show()
```

```python
top_categories = category_revenue.head(10).index
pivot_top = df[df['product_category_name_english'].isin(top_categories)] \
    .pivot_table(
        values='price',
        index='product_category_name_english',
        columns='Month',
        aggfunc='sum',
        fill_value=0
    )
plt.figure(figsize=(12, 6))
sns.heatmap(
    pivot_top,
    cmap="YlGnBu",
    annot=True,
    fmt=".0f"
)
plt.title("Top 10 Categories Revenue by Month")
plt.xlabel("Month")
plt.ylabel("Product Category")
plt.show()
```

```python
total_revenue = df['price'].sum()
print(total_revenue)
```

```python
total_orders = df['order_id'].nunique()
print(total_orders)
```

```python

best_month = str(monthly_sales.idxmax().to_timestamp().strftime("%B %Y"))
print(best_month)
```

```python
print("Total Revenue :", round(total_revenue,2))
print("Total Orders :", total_orders)
print("Top Category :", top_category)
print("Best Month :", best_month)
```
